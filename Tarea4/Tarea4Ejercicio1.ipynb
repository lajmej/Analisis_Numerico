```python
#Ejercicio 1.1 Bisección

# Definir la funcion
def f(x):
    return x**3-5*x-9

# Metodo de Bisección
def bisection(x0,x1,e):
    step = 1
    print('\n\n*** Implementar Metodo ***')
    condition = True
    while condition:
        x2 = (x0 + x1)/2
        print('Iterasion-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2

        step = step + 1
        condition = abs(f(x2)) > e

    print('\nRaiz requerida : %0.8f' % x2)


# Input
x0 = input('Primer valor: ')
x1 = input('Segundo valor: ')
e = input(' Error Tolerable : ')

#  input a float
x0 = float(x0)
x1 = float(x1)
e = float(e)



# Checar los valores iniciales
if f(x0) * f(x1) > 0.0:
    print('Los valores no dan soporte a las raices.')
    print('Prueba otros valores.')
else:
    bisection(x0,x1,e)
```

```python
#1.2 Secante
# Definir la funcion
def f(x):
    return x**3 - 5*x - 9

# Metodo Secante

def secant(x0,x1,e,N):
    print('\n\n*** Implementar metodo ***')
    step = 1
    condition = True
    while condition:
        if f(x0) == f(x1):
            print('Error por dividir entre cero!')
            break

        x2 = x0 - (x1-x0)*f(x0)/( f(x1) - f(x0) )
        print('Iterasion-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))
        x0 = x1
        x1 = x2
        step = step + 1

        if step > N:
            print('No converge !')
            break

        condition = abs(f(x2)) > e
    print('\n Raiz requerida: %0.8f' % x2)


# Input Section
x0 = input('primer valor: ')
x1 = input('segundo valor: ')
e = input(' Error Tolerable: ')
N = input('Paso máximo: ')

# convertir x0 y e a float
x0 = float(x0)
x1 = float(x1)
e = float(e)

# convertir N a integer
N = int(N)


# Empezar metodo secante
secant(x0,x1,e,N)
```

```python
#1.3 Metodo Newton
# Definir funcion
def f(x):
    return x**3 - 5*x - 9

# Definir derivada de la funcion
def g(x):
    return 3*x**2 - 5

# Implenetar Newton Rapson

def newtonRaphson(x0,e,N):
    print('\n\n*** Implementar Metodo ***')
    step = 1
    flag = 1
    condition = True
    while condition:
        if g(x0) == 0.0:
            print('Error por dividir entre cero!')
            break

        x1 = x0 - f(x0)/g(x0)
        print('Iterasion-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
        x0 = x1
        step = step + 1

        if step > N:
            flag = 0
            break

        condition = abs(f(x1)) > e

    if flag==1:
        print('\nRaiz requerida: %0.8f' % x1)
    else:
        print('\nNo converge.')


# Input Section
x0 = input('pon el primer valor de x0: ')
e = input('error tolerable : ')
N = input('pasos máximos: ')

# Converting x0 and e to float
x0 = float(x0)
e = float(e)

# Converting N to integer
N = int(N)


# Starting Newton Raphson Method
newtonRaphson(x0,e,N)
```
