```python
#Ejercicio 7
import numpy as np
import matplotlib.pyplot as plt

# Definir la función g(x)
def g(x):
    return np.pi + 0.5 * np.sin(x / 2)

# Parámetros de la iteración de punto fijo
tolerance = 1e-2
max_iter = 1000

# Punto inicial
x0 = 0

# Realizar la iteración de punto fijo
x_values = [x0]
for _ in range(max_iter):
    x_new = g(x0)
    x_values.append(x_new)
    if abs(x_new - x0) < tolerance:
        break
    x0 = x_new

# Punto fijo encontrado
fixed_point = x_values[-1]
print(f'Punto fijo: {fixed_point}')

# Graficar la función g(x) y la iteración
x = np.linspace(0, 2 * np.pi, 400)
y = g(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$g(x) = \pi + 0.5 \sin(x/2)$')
plt.plot(x, x, 'r--', label='$y = x$')
plt.scatter(x_values, g(np.array(x_values)), color='blue', marker='o', label='Iteraciones')
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.title('Iteración de Punto Fijo para $g(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
